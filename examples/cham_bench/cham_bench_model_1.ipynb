# Step functions

```python
# cham_bench loop, for the step functions
cmf_grid = pd.read_csv('/Users/honeyeah/Codes/coregiver/examples/cham_bench/cham_bench_cmf_grid.csv')
bounds = cmf_grid['bounds']
percentages = cmf_grid['percentages']
n_runs = 30 # This variable is not used in the loop, you might want to remove it
for i in range(n_runs):
    for _, row in cmf_grid.iterrows():
        bound = row['bounds']
        perc = row['percentages']

        collision_file = f"/Users/honeyeah/Codes/data/cham_bench/cham_bench_{i+1}/collision_report.txt"
        comp_input_file = f"/Users/honeyeah/Codes/coregiver/examples/cham_bench/step_functions/cham_bench_step_b{bound}_p{perc}.txt"

        ejec_file = f"/Users/honeyeah/Codes/data/cham_bench/cham_bench_{i+1}/ejections.txt"

        if os.path.exists(ejec_file):
            print('ejections file found.', i)
        else:
            ejec_file = "/Users/honeyeah/Codes/planetgarten/body_composition_tracker/test_input/ejections_empty.txt"
            print("ejections file not found, using an empty file.", i)

        comp_output_file = f"/Users/honeyeah/Codes/data/cham_bench/cham_bench_{i+1}/cham_bench_output_step_b{bound}_p{perc}.txt"

        final_compositions = ctv1.track_composition(collision_file, comp_input_file, ejec_file)
        ctv1.write_output(final_compositions, comp_output_file)
        print(f"Processed cham_bench run {i+1}")
```
